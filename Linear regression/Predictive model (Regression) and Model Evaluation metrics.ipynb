```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Transformed_Housing_Data_2-Copy1')
data.head()
```

```python
data = data.rename(columns={'Sale_Price_x':'Sale_Price'})
```

```python
data.head()
```

```python
data['mean_sales'] = data['Sale_Price'].mean()
data['mean_sales'].head()
```

```python
#resolution of image
plt.figure(dpi=100)

#list k : index for data (x axis)
k=range(0,len(data))

plt.scatter(k,data['Sale_Price'].sort_values(),color = 'blue',label='Actual Sale Price')
plt.plot(k,data['mean_sales'].sort_values(),color ='green',label = 'mean-price')
plt.xlabel('Fitted pointd (Ascending)')
plt.ylabel('Sale Price')
plt.title('Overall Mean')
plt.legend()
```

```python
# average sale price with respect to overall grade
grades_mean = data.pivot_table(values = 'Sale_Price',index= 'Overall Grade',aggfunc = np.mean)
grades_mean
```

```python
# we can use any categorical variable , muliple of then to get mean of sale price
# 10 diff value - 10 unique value for overall grade
grades_mean = data.pivot_table(values = 'Sale_Price',columns = 'Overall Grade',aggfunc = np.mean)
grades_mean
```

```python
bedrooms_mean = data.pivot_table(values = 'Sale_Price',columns = 'No of Bedrooms',aggfunc = np.mean)
bedrooms_mean
```

```python
Flat_Area_mean = data.pivot_table(values = 'Sale_Price',columns = 'Flat Area (in Sqft)',aggfunc = np.mean)
Flat_Area_mean
```

```python
data['Overall Grade'].head()
```

```python
#creater new column
data['grade_mean']=0

#for every grade full its mean price in new column
for i in grades_mean.columns:
    data['grade_mean'][data['Overall Grade']== i] = grades_mean[i][0]
    
data['grade_mean'].head(16)
```

```python
data['Sale_Price'][data['Overall Grade']==1]
```

```python
#compare with previous model
gradewise_list = []
for i in range(1,11):
    k=data['Sale_Price'][data['Overall Grade']==i]
    gradewise_list.append(k)

for i in range(0, len(gradewise_list)):

     print(gradewise_list[i])
```

```python
classwise_list=[]
for i in range(1,11):
    k=data['Sale_Price'][data['Overall Grade']==i]
    classwise_list.append(k)
```

```python
plt.figure(dpi=120,figsize=(15,9))
#ploting sale_price gradewise 
#z variable is for x-axis
z=0
for i in range(1,11):
    #defining x-axis using z
    points = [k for k in range(z,z+ len(classwise_list[i-1]))]
    #plotting
    plt.scatter(points,
               classwise_list[i-1].sort_values(),
               label = ('houses with overall grade',i),s=4)
    #plotting gradewise mean
    plt.scatter(points,
                [classwise_list[i-1].mean() for q in range(len(classwise_list[i-1]))],
                s=6, color ='pink'
               )
    z = max(points) + 1
    
  #plotting overall mean
plt.scatter([q for q in range(0,z)],
               data['mean_sales'],
               color='red',
               label ='Overall mean',
               s=6)
plt.xlabel('Fitted points (Ascending)')
plt.ylabel('Sale Price')
plt.title('Overall Mean')
plt.legend(loc=4) #position of legend 

    
```

```python
#residual plt

mean_difference = data['mean_sales']-data['Sale_Price']#predictied 
grade_mean_difference = data['grade_mean']-data['Sale_Price']
```

```python
# list of indices for datapts
k = range(0,len(data))

#list of 0 , represent residual of a perfect model where prediction same as actual hence residual 0
l=[0 for i in range(len(data))]


plt.figure( figsize=(15,6),dpi=100)

plt.subplot(1,2,1)

#residual for overall mean sale price
plt.scatter(k,mean_difference,color='red',label='Residuals',s=2)
plt.plot(k,1,color='green',label='mean Regression',linewidth=3)
plt.xlabel('Fitted points')
plt.ylabel('Residuals')
plt.title('Resuidals with respect to Gradewise Mean')

plt.subplot(1,2,2)
#residual  mean sale price wrt to overall grade
plt.scatter(k,grade_mean_difference, color='red',label='Residuals',s=2)
plt.plot(k,1,color='green',label='mean Regression',linewidth = 3)
plt.xlabel('Fitted points')
plt.ylabel('Residuals')
plt.legend()
plt.title('Resuidals with respect to Gradewise Mean')

plt.legend()
```
